### Penghapusan nilai unique / tidak valid

Ketika dilakukan pengecekan ada inputan yang nilainya "?", maka dalam kasus ini akan kita hapus

```python
column = ["Gender", "CALC", "FAVC",	"SCC",	"SMOKE",	"family_history_with_overweight",	"CAEC", "MTRANS",	"NObeyesdad"]
for col in column:
    print(f"Kolom: {col}")
    print(df[col].unique())
    print("-" * 40)
```

```python
# Ganti '?' menjadi np.nan untuk kolom-kolom kategori
column = ["Gender", "CALC", "FAVC", "SCC", "SMOKE", 
          "family_history_with_overweight", "CAEC", "MTRANS", "NObeyesdad"]

for col in column:
    df[col] = df[col].replace('?', np.nan)

# Cek jumlah missing value setelah dibersihkan
print(df[column].isnull().sum())
```

### SimpleImputer  
untuk menangani Missing Value, menggunakan Median dan Mean untuk mengganti data yang hilang.

```python
num_cols = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']
cat_cols = ['Gender', 'CALC', 'FAVC', 'SCC', 'SMOKE',
            'family_history_with_overweight', 'CAEC', 'MTRANS']
```

```python
# 1.c  Imputasi:   median utk numerik, modus utk kategori
num_imputer  = SimpleImputer(strategy='median')
cat_imputer  = SimpleImputer(strategy='most_frequent')
```

```python
df[num_cols] = num_imputer.fit_transform(df[num_cols])
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])
```

```python
df.isnull().sum()
```

### Winsorizing untuk Outlier
untuk menangani outlier, Mengurangi pengaruh outlier ekstrem pada model, tanpa harus menghapus baris data tersebut.

```python
for col in num_cols:
    q1, q3 = df[col].quantile([0.25, 0.75])
    iqr    = q3 - q1
    lower  = q1 - 1.5 * iqr
    upper  = q3 + 1.5 * iqr
    df[col] = np.clip(df[col], lower, upper)
```

Split pertama (sebelum scaling/SMOTE):

Menghindari data leakage saat scaling atau SMOTE.

Misalnya: hanya melakukan scaling dan SMOTE pada data latih (agar informasi dari data uji tidak bocor).

```python
X = df.drop('NObeyesdad', axis=1)
y = df['NObeyesdad']
```

```python
# 1. Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# 2. Scaling
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Tampilkan beberapa data hasil scaling (train)
print("Contoh data X_train setelah scaling:")
print(pd.DataFrame(X_train_scaled, columns=X.columns).head())
```

```python
# 3. SMOTE pada data train
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train_scaled, y_train)
```

```python
# Tampilkan ukuran sebelum dan sesudah SMOTE
print(f"Ukuran data train sebelum SMOTE: {X_train.shape}, label distribusi:\n{y_train.value_counts()}")
print(f"Ukuran data train setelah SMOTE: {X_train_res.shape}, label distribusi:\n{pd.Series(y_train_res).value_counts()}")
```

```python
# Visualisasi distribusi label sebelum dan sesudah SMOTE
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Sebelum SMOTE
sns.countplot(x=y_train, ax=axes[0])
axes[0].set_title('Distribusi Label Sebelum SMOTE')
axes[0].tick_params(axis='x', rotation=45)

# Sesudah SMOTE
sns.countplot(x=y_train_res, ax=axes[1])
axes[1].set_title('Distribusi Label Sesudah SMOTE')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```


Menggabungkan hasil dari SMOTE ke data frame awal 

```python
# Contoh data hasil SMOTE (train)
df_train_res = pd.DataFrame(X_train_res, columns=X.columns)
df_train_res['NObeyesdad'] = y_train_res.values
print("Contoh data train setelah SMOTE:")

df_train_res.head(20)
```
